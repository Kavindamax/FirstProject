```python
country_score = data.groupby('Country name')['Ladder score'].sum().sort_values(ascending=False).head(50)
```

```python
country_score
```

```python
plt.figure(figsize=(20,20))
fig=sns.barplot(x=country_score.index, y=country_score, data=country_score, color = 'red')
fig.set_xticklabels(rotation=90, labels=country_score.index)
plt.xlabel('country')
fig
```

```python
plt.figure(figsize=(20,10))
fig = sns.boxplot(data['Logged GDP per capita'])
```

```python
fig = sns.distplot(data['Logged GDP per capita'], color = 'blue')
```

```python
fig = sns.relplot(
      data = data,
      x = 'Healthy life expectancy', y = 'Ladder score',
      hue = 'Regional indicator',
      palette = 'magma', sizes=(30,400),)
```

```python
plt.figure(figsize=(20,10))
sns.lmplot(x='Explained by: Perceptions of corruption',
           y='Perceptions of corruption',
           data=data,
           fit_reg=False,
           hue='Regional indicator',
           legend=False,)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(data.corr(), cmap="RdYlBu", annot=True, vmin=-1, vmax=1, linewidths=0.5, center=0)
```
